```python
import pandas as pd
df = pd.read_json (r'C:\Users\ricky\Downloads\DL_final\result-10.json')
```

```python
df.size
```

```python
# indexs = ["center_x", "center_y" , "width", "height"]
data = []

import os
from PIL import Image

for i in range(6744):
    file_name = df['filename'][i].split("/")
    path = os.path.join(os.getcwd(),file_name[5])
    img = Image.open(path)
    print(img.size)

    one_name = str(df['filename'][i])
    one_name = one_name.split("/")[-1]
    one_obj = df['objects'][i]

    for one_box in one_obj:
        tmp = []
        tmp.append(one_name)
        tmp.append(one_box["class_id"])
        w = abs(one_box["relative_coordinates"]["width"]*img.size[0])
        h = abs(one_box["relative_coordinates"]["height"]*img.size[1])
        tmp.append(abs((one_box["relative_coordinates"]["center_x"]*img.size[0]) - (w/2)))
        tmp.append(abs((one_box["relative_coordinates"]["center_y"]*img.size[1]) - (h/2)))
        tmp.append(w)
        tmp.append(h)
        tmp.append(one_box["confidence"])
        data.append(tmp)
```

```python
data = pd.DataFrame(data)
data.columns = ["image_filename", "label_id", "x", "y", "w", "h", "confidence"]
data
```

```python
data.to_csv (r'C:\Users\ricky\Downloads\result.csv', index = None)
```
